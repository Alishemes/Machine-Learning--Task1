```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data = pd.read_csv("salaries(1).csv")
data
```

```python
data.head(10)
```

```python
data.tail(5)
```

```python
input = data.drop('has_diabetes', axis='columns')
input
```

```python
target = data['has_diabetes']
target
```

```python
la_age = LabelEncoder()
la_bmi = LabelEncoder()
la_family = LabelEncoder()
la_glucose = LabelEncoder()
```

```python
input['age_group_n'] = la_age.fit_transform(input['age_group'])
input['bmi_category_n'] = la_bmi.fit_transform(input['bmi_category'])
input['family_history_n'] = la_family.fit_transform(input['family_history'])
input['glucose_level_n'] = la_glucose.fit_transform(input['glucose_level'])
```

```python
input
```

```python
input_n = input.drop(['age_group', 'bmi_category', 'family_history', 'glucose_level'], axis='columns')
input_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(input_n, target)
```

```python
input_data = [[2, 1, 0, 1]]
input_df = pd.DataFrame(input_data, columns=['age_group_n', 'bmi_category_n', 'family_history_n', 'glucose_level_n'])

prediction = model.predict(input_df)
print(prediction)
```

```python
input_data = [[0, 2, 1, 0]]
input_df = pd.DataFrame(input_data, columns=['age_group_n', 'bmi_category_n', 'family_history_n', 'glucose_level_n'])

prediction = model.predict(input_df)
print(prediction)
```
